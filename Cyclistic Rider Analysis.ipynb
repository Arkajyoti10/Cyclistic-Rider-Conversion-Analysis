# Feature Engineering

## Datatype handling for further calculation

```python
df.dtypes
```

```python
df['started_at']=pd.to_datetime(df['started_at'], format='mixed')
df['ended_at']=pd.to_datetime(df['ended_at'], format='mixed')
df['ride_length']=pd.to_timedelta(df['ride_length'])
df['ride_length']=df['ride_length'].dt.total_seconds() / 60
```

```python
df.dtypes
```

## Added hour of week column to detect peak riding hours

```python
df['hour_of_day']=df['started_at'].dt.hour
```

```python
df.head()
```

```python
df.dtypes
```

# Data Analysis

## Trip duration: Casual Vs Member

```python
ride_length_summary=df.groupby('member_casual')['ride_length'].agg(['mean','max','median'])
```

```python
ride_length_summary
```

```python
avg_rides=df.groupby('member_casual')['ride_length'].mean().reset_index()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=avg_rides, x='member_casual', y='ride_length', palette='viridis')

plt.title('Average Trip duration: Casual Vs Member', fontsize=14)
plt.xlabel('Rider type', fontsize=12)
plt.ylabel('Average duration', fontsize=12)

plt.show()
```

#### Casual riders have a higher average riding duration (22 mins) which suggests they use bikes for leisure on the other hand member riders have a lower duration (12 mins) which confirms they use bikes for commuter purposes

## Hour-wise Rides analysis

```python
rider_counts=df.groupby(['hour_of_day','member_casual']).size().reset_index(name='ride_id')
hourly_summary=rider_counts.groupby(['hour_of_day','member_casual'])['ride_id'].mean().reset_index()
casual_data=hourly_summary[hourly_summary['member_casual']=='casual']
member_data=hourly_summary[hourly_summary['member_casual']=='member']
```

```python
plt.figure(figsize=(10,6))
plt.plot(casual_data['hour_of_day'], casual_data['ride_id'],label='Casual',color='red',marker='^',linewidth=2)
plt.plot(member_data['hour_of_day'], member_data['ride_id'],label='Member',color='blue',marker='o',linewidth=2)

plt.xlabel('Hour-wise ', fontsize=12)
plt.ylabel('Average number of rides taken', fontsize=12)
plt.title('Hour-wise rider Analysis', fontsize=14)
plt.legend(title='Rider type')
plt.xticks(range(0,24))
plt.show()
```

### The peak hours for member riders are 8 AM and 5 PM (17:00) over Casual riders who peak around 3 PM-5 PM which proves member users are generally commuters

```python
#creating a weekend condition if casual rides are high in 6=friday,7=saturday anmd 1=sunday
df['day_type']=df['days_of_week'].apply(lambda x:'Weekend' if x in [6,7,1] else 'Weekday')
hourly_segment=df.groupby(['day_type','hour_of_day','member_casual']).size().reset_index(name='count')
```

```python
sns.relplot(data=hourly_segment,x='hour_of_day',y='count',hue='member_casual',col='day_type',kind='line')
```

```python
rider_mix=df.groupby(['day_type','member_casual']).size().unstack()
rider_mix_pct=rider_mix.divide(rider_mix.sum(axis=1),axis=0)*100
rider_mix_pct=rider_mix_pct.round(2)
rider_mix_pct
```

### The graph clearly shows that there is decrease in number of rides in weekends but the data provided suggests Casual members contribute a higher amount of riders during the weekends which proves casual riders are weekend warriors

## Weekly Trip duration Analysis

### For the given analysis 1=Sunday and 7=Saturday

```python
df=df[df['days_of_week'].isin([1,2,3,4,5,6,7])]
weekly_summary=df.groupby(['days_of_week', 'member_casual'])['ride_length'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(data=weekly_summary,x='days_of_week',y='ride_length',hue='member_casual',palette='Set2')

plt.title('Weekly trip duration analysis', fontsize=14)
plt.xlabel('Day of the week', fontsize=12)
plt.ylabel('Average duration', fontsize=12)
plt.legend(title='Rider type')

plt.show()
```

### The data clearly shows the average trip durations for member users almost remains the same all through the week but Casual riders mostly ride longer during the weekends(1=sunday and 5=Friday)

## Average Monthly rider Analysis

### For this Analysis 1=January and 12=December

```python
df['month_of_ride']=df['started_at'].dt.month
df.head()
```

```python
monthly_summary=df.groupby(['month_of_ride', 'member_casual'])['ride_length'].mean().reset_index()
casual_data=monthly_summary[monthly_summary['member_casual']=='casual']
member_data=monthly_summary[monthly_summary['member_casual']=='member']
```

```python
plt.figure(figsize=(10,6))
plt.bar(casual_data['month_of_ride'], casual_data['ride_length'],label='Casual',color='skyblue')
plt.plot(member_data['month_of_ride'], member_data['ride_length'],label='Member',color='darkorange',marker='o',linewidth=2)

plt.xlabel('Months', fontsize=12)
plt.ylabel('Average trip duration', fontsize=12)
plt.title('Average monthly rider Analysis', fontsize=14)
plt.legend(title='Rider type')
plt.xticks(range(1,13))
plt.show()
```

### Casual rides peak during the summer seansons(5=May to August=8) over member riders who almost remain consistent over all the seasons shown by the bell curve

# Peak riding Stations

```python
casual_rides=df[df['member_casual']=='casual']
top_10_casual=casual_rides['start_station_name'].value_counts().head(10).reset_index()
top_10_casual.columns=['start_station_name','ride_id']

plt.figure(figsize=(10,6))
sns.barplot(data=top_10_casual,x='ride_id',y='start_station_name',palette='viridis')
plt.title('Top 10 stations for Casual riders', fontsize=14)
plt.ylabel('Station names',fontsize=12)
plt.show()
```

### The data clearly points out the top 10 stations for casual riders where most people start their journey, these are our 'hubs'
